### Which one should i use? (pyplot or matplotlib oo method)
* When plotting something quickly use pyplot method
* When plotting something more advanced use oo method

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO method 
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
#ax.set_xlim(45,100)
```

```python
#OO method from the scratch
fig, ax = plt.subplots(figsize = (10,6))
#PLot the data 
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target']);
#Customize the plot
ax.set(title= 'Heart Disease and Cholestrol',
      xlabel='Age',
      ylabel='Cholestrol');
#add a legend
ax.legend(*scatter.legend_elements(), title="Target")
#Add the horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle='--')
```

```python
over_50.target.values
```

```python
#subplot of age,chol and thalach
fig , (ax0, ax1) = plt.subplots(figsize=(10,10),
                               nrows=2,
                               ncols=1,
                               sharex=True)

#add a data to ax0
scatter =ax0.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target']);

#customize the ax0
ax0.set(title = 'Heart Disease and Cholestrol Levels',
       ylabel='Cholestrol');

#add a legend 
ax0.legend(*scatter.legend_elements(), title='Target');

#Add a main line
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--');

#Add data to ax1
scatter =ax1.scatter(x=over_50['age'],
                    y=over_50['thalach'],
                    c=over_50['target']);


#customize the ax1
ax1.set(title = 'Heart Disease and Max Heart rate',
       xlabel='Age',
       ylabel='Max heart rate ');

#add a legend 
ax1.legend(*scatter.legend_elements(), title='Target');

#Add a main line
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='--');

#Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16,fontweight='bold');
```
